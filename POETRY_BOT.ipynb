```python
import cv2
import pytesseract
import os
from IPython.display import clear_output
import markovify
```

```python
os.system('cd Image-Downloader; python3 image_downloader.py --engine "Google" --driver chrome --max-number 500 --output "POEMS" "philosophical poems"')
```

```python
img_files = os.popen('ls /Users/hunter/Desktop/hacking/Image-Downloader/POEMS').read()
img_files = img_files.split('\n')
```

```python
print(img_files)
```

```python
totalString = ""
total = len(img_files)
for file in img_files:
    if total == 2:
        break
    img_cv = cv2.imread('/Users/hunter/Desktop/hacking/Image-Downloader/POEMS/' + file)
    img_rgb = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)
    candidate = pytesseract.image_to_string(img_rgb).lower()
    if len(candidate.split(' ')) > 5:
        totalString += candidate
    clear_output(wait=True)
    print(str(total) + " files remaining...")
    total = total - 1 
clear_output(wait=True)
print("done!")
```

```python
print(totalString)
```

```python
text_model = markovify.Text(totalString)
```

```python
for i in range(1000):
    print(str(i) + ': ')
    print(text_model.make_sentence(), '\n')
```
